# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore')
```

```python
df.head()
```

```python
# Distributions: 
df1 = sns.distplot(df["InternetUsers"])
plt.show()
```

```python
df1 = sns.displot(df["InternetUsers"], bins=10)
plt.show()
```

```python
#BOX PLOTS:
df2 = sns.boxplot(data = df, x="IncomeGroup", y='BirthRate')
plt.show()
```

```python
df3 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate', fit_reg = False) #lm - linear model
plt.show()
```

```python
df4 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate')
plt.show()
```

```python
df5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
plt.show()
```

```python
df5 = sns.lmplot(data = df,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup', height = 10)
plt.show()
```
